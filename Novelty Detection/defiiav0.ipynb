```python
def cntpic(o):
    l =[]
    val = []
    for k in train.index:
        value = 0
        s =0
        for j in train.iloc[k].values:
            if abs(j) > (abs(np.mean(train.iloc[k].values)) + o*np.std(train.iloc[k].values)):
                value = value + (abs(j) - (abs(np.mean(train.iloc[k].values)) + o*np.std(train.iloc[k].values)))
                s = s+1
        val.append(value)
        l.append(s)
    return(l,val)
```

```python
#fenetrage
def dif_corr(df):
    fn = np.arange(0,61440,1024)
    mat = []
    pval=[]
    for k in df.index: 
        l = []
        val = []
        for j in fn:
            cr = corr(train.iloc[k].values[j:j+1024],np.arange(len(train.iloc[k].values[j:j+1024])))
            l.append(abs(cr[0]) )
            val.append(cr[1])
        mat.append([np.max(l),np.min(l), np.max(l)-np.min(l)])
        pval.append([np.max(val),np.min(val)])
    return(mat,pval)
```

```python
def getbar(df):
    lmoy = []
    lsgm = []
    for k in df.index:
        lmoy.append(np.mean(df.iloc[k].values))
        lsgm.append(np.std(df.iloc[k].values))
    return([np.mean(lmoy),np.mean(lsgm)])
```

```python
from scipy.spatial import distance
def getdistbar(df,bar):
    dist = []
    for k in df.index:
        dist.append(distance.cityblock([np.mean(df.iloc[k].values),np.std(df.iloc[k].values)],bar))
    return(dist)
        
    
```
